### 2x + 4y = 6
### -2x + y = 4

### Importerer nødvendige biblotek fra Python

```python
import matplotlib.pyplot as plt 
import numpy as np
import sympy as sp
```

```python
#### finn verdiene for X og Y
x , y=sp.symbols("x y")
eq1=sp.Eq(2*x+4*y,6)
eq2=sp.Eq(-2*x+y, 4)

sp.solve((eq1,eq2),(x,y))
```
